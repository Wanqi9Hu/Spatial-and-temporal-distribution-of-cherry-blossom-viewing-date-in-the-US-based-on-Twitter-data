```python
tweets_df['Content'] = tweets_df.Content.str.lower()
tweets_df
```

```python
target=['cafe','perfume','noodle','dance studio','lager', 'hair', 'studio','drinking','wine','banquet hall','bar','ale','gift','cocktail','spa','golf','club','cupcake','cream','starbuck','#starbuck#','parade','run','#starbuck','drink','score','candy','delicious','running','tea','head','band','milk','cola','artechouse','']
```

```python
df = []
for i in range(0,tweets_df.shape[0]):
    temp=tweets_df.iloc[i,0].split()
    justify=True
    for m in temp:
        if m in target:
            justify=False
            break
    if justify:
        df.append([tweets_df.iloc[i,0],tweets_df.iloc[i,2], tweets_df.iloc[i,3], tweets_df.iloc[i,4], tweets_df.iloc[i,5]])
df = pd.DataFrame(df, columns=['Content','state','year','month','day'])
df
```

```python
date=[]
for i in range(0,df.shape[0]):
    if len(str(df.iloc[i,4]))==1:
        temp=str(df.iloc[i,3])+'/'+'0'+str(df.iloc[i,4])
    else:
        temp=str(df.iloc[i,3])+'/'+str(df.iloc[i,4])
    date.append(temp)
df['date']=date
df
```

```python
open_date={}
key_1=df['state'].unique().tolist()
key_2=df['year'].unique().tolist()
key_3=df['date'].unique().tolist()
```

```python
key_3.sort()
```

```python
for i in key_1:
    open_date[i]={}
    for j in key_2:
        open_date[i][j]={}
        for n in key_3:
            open_date[i][j][n]=0
```

```python
for i in range(0,df.shape[0]):
    open_date[df.iloc[i,1]][df.iloc[i,2]][df.iloc[i,5]]+=1
open_date
```

```python
exact_date={}
for i in key_1:
    exact_date[i]={}
    for j in [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]:
        exact_date[i][j]=''
```

```python
for i in open_date.keys():
    for j in open_date[i].keys():
        all=open_date[i][j]
        time='3/1'
        for m,n in all.items():
            if m>time and n>=1:
                time=m
                break
        exact_date[i][j]=time
```

```python
cherry_blossom=[]
for i in exact_date.keys():
    year=[]
    for m,n in exact_date[i].items():
        year.append(n)
    cherry_blossom.append(year)
cherry_blossom = pd.DataFrame(cherry_blossom, columns=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021])
cherry_blossom['state']=exact_date.keys()
cherry_blossom
```

```python
tweet=[]
for i in range(0,cherry_blossom.shape[0]):
    temp=[]
    for j in range(0,11):
        year=j+2011
        daytime=cherry_blossom.iloc[i,j]
        loc=cherry_blossom.iloc[i,11]
        huayu=None
        for m in range(0,df.shape[0]):
            if (df.iloc[m,1]==loc and df.iloc[m,2]==year) and (df.iloc[m,5]==daytime):
                huayu=df.iloc[m,0]
                break
        temp.append(huayu)
    tweet.append(temp)
tweet= pd.DataFrame(tweet, columns=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021])
tweet['state']=exact_date.keys()
tweet
```
